```python
import docx2txt
```

```python
path="testdoc1.docx"
text1=docx2txt.process(path)
print(text1)
```

```python
path="testdoc2.docx"
text2=docx2txt.process(path)
print(text2)
```

```python
#tokenization for sentence
from nltk.tokenize import sent_tokenize
sent1=sent_tokenize(text1)
sent2=sent_tokenize(text2)
print(sent1,sent2)
```

```python
#tokenization of word
from nltk.tokenize import word_tokenize
word1=word_tokenize(text1)
word2=word_tokenize(text2)
print(word1,word2)
```

```python
#pos tagging
import nltk
nltk.pos_tag(word1)
```

```python
#checking frequency
from nltk.probability import FreqDist
freq1=FreqDist(word1)
freq2=FreqDist(word2)
print(freq1)
print(freq2)
```

```python
freq1.most_common()
```

```python
freq2.most_common()
```

```python
#show no of words
import matplotlib.pyplot as plt
freq1.plot(20)
plt.show()
```

```python
#removal of stop words
from nltk.corpus import stopwords
stop_words=set(stopwords.words("english"))
print(stop_words)
```

```python
#new list without stop words
new_word=[]
for i in word1:
    if i not in stop_words:
        new_word.append(i)
print(new_word)        
```

```python
#stemming
from nltk.stem import PorterStemmer
ps=PorterStemmer()
stemmed_word=[]
for i in new_word:
    stemmed_word.append(ps.stem(i))
print("Old Words")
print(new_word)
print("Stemmed words")
print(stemmed_word)
```

```python
#lemmatization
from nltk.stem.wordnet import WordNetLemmatizer
wnl=WordNetLemmatizer()
lemmatize_word=[]
for i in new_word:
    lemmatize_word.append(wnl.lemmatize(i))
print("Old word")
print(new_word)
print("Lemmatized_word")
print(lemmatize_word)
```

```python
#Calculating term frequency
documents=[text1,text2]
for i in documents:
    dict={}
    for word in i.split():
        if word in dict:
            dict[word]+=1
        else:
            dict[word]=1
            
            
    for key,value in dict.items():
        dict[key]=value/len(i)
        print(key,dict[key])
```
