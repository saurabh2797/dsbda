```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
df=pd.read_csv('nba.csv')
df
```

```python
print("Mean of Age : ",df['Age'].mean())
print("Median of Age : ",df['Age'].median())
print("Max of Age : ",df['Age'].max())
print("Min of Age : ",df['Age'].min())
print("STD of Age : ",df['Age'].std())
```

```python
print("Mean of Salary : ",df['Salary'].mean())
print("Median of Salary : ",df['Salary'].median())
print("Max of Salary : ",df['Salary'].max())
print("Min of Salary : ",df['Salary'].min())
print("STD of Salary : ",df['Salary'].std())
```

```python
df.dtypes
```

```python
#Grouping by categorical variable - Age
bins=[19,25,31,36,40]
labels = ['19-24','25-30','31-35','36-40']
df['AgeGroup'] = pd.cut(df['Age'],bins=bins,labels=labels)
```

```python
df
```

```python
df.groupby('AgeGroup')['Salary'].mean().describe()
```

```python
salarylist=list(df.groupby('AgeGroup')['Salary'])
```

```python
salarylist
```

```python
pivot_table = df.pivot_table(values='Age',index='Salary',columns=df['AgeGroup'],aggfunc=['mean','median','max','min','std'])
pivot_table
```

```python
crosstab = pd.crosstab(index=df['Salary'],values=df['Salary'],columns=df['AgeGroup'],aggfunc=['mean','median','max','min','std'])
crosstab
```
